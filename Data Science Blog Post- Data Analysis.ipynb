```python
#create copy of df to fix datatypes
goodread_copy_df = goodread_df.copy()
goodread_copy_df.head()

goodread_copy_df.dtypes

#update publication_date to datetime
goodread_copy_df['publication_date'] = pd.to_datetime(goodread_copy_df['publication_date'])
  
goodread_copy_df.dtypes
```

```python
#create Boolean series for publishers that contain the phrase .com
digitalselfpublisher_bool_series = goodread_copy_df['publisher'].str.contains('.com')

#create audiobook df slice
digitalselfpublisher_bool_slice = digitalselfpublisher_bool_series[(digitalselfpublisher_bool_series == True)]

#create a list of indexes for audiobooks
digitalselfpublisher_bool_index = list(digitalselfpublisher_bool_slice.index.values)

#create a new df with audiobook rows
df_digitalselfpublisher = goodread_copy_df.loc[digitalselfpublisher_bool_index]
df_digitalselfpublisher.head()
```
